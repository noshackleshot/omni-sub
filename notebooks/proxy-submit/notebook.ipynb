# OmniSub Proxy Submit
CPU-only notebook that copies a precomputed submission.csv to output.

```python
import os, shutil, glob

# Find precomputed submission
src = None
for pattern in [
    '/kaggle/input/omnisub-precomputed/submission.csv',
    '/kaggle/input/omnisub-precomputed/*.csv',
]:
    matches = glob.glob(pattern)
    if matches:
        src = matches[0]
        break

if not src:
    # List available input files for debugging
    print('Available inputs:')
    for root, dirs, files in os.walk('/kaggle/input'):
        for f in files:
            print(f'  {os.path.join(root, f)}')
    raise FileNotFoundError('No submission.csv found in omnisub-precomputed dataset')

dst = '/kaggle/working/submission.csv'
shutil.copy(src, dst)

# Validate
import csv
with open(dst) as f:
    reader = csv.reader(f)
    header = next(reader)
    rows = sum(1 for _ in reader)

print(f'Source: {src}')
print(f'Header: {header}')
print(f'Rows: {rows}')
print(f'Output: {dst} ({os.path.getsize(dst)} bytes)')
assert header == ['path', 'transcription'], f'Bad header: {header}'
assert rows == 3000, f'Expected 3000 rows, got {rows}'
print('Submission ready!')
```
